# BS tutorial
In this tutorial I am going to parse a website to gather some useful information

```python
from bs4 import BeautifulSoup
import requests
import lxml
```

```python
# Send an HTTP GET request to the website and store the response in 'result'
#website = "https://subslikescript.com/movie/Titanic-120338"
website = "https://meisamghasemi.com"
result = requests.get(website)

# Extract the raw HTML content from the response as a string
content = result.text

# Parse the HTML content using BeautifulSoup with the built-in HTML parser
#soup = BeautifulSoup(content, "html.parser")
soup = BeautifulSoup(content, "lxml")
print(soup.prettify())
```

```python
print(soup.find_all("h4"))
print(soup.findAll("p", class_="elementor-image-box-description"))
obj = soup.findAll("p", class_="elementor-image-box-description")
[print(object.get_text(strip=True)) for object in obj]
print(soup.find("h1").get_text(strip=True))
```

```python
with open("Meisam_site.txt", "w", encoding="utf-8") as file:
    for object in obj:
        text = object.get_text(strip=True)
        file.write(text)
```
